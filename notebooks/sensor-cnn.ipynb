```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from scipy.ndimage import gaussian_filter1d
from collections import defaultdict

def draw_single_trial(category, trial=1, trend_threshold=0.004, variance_threshold=0.000008):
    prep = Preprocessor()
    input_data = prep._load_input_data(category)

    normalized_data = prep._normalize_data(input_data, True, True, True)
    original_data = prep._normalize_data(input_data, True, True, False)    

    prep._relevant_channels_filtering(normalized_data, trend_threshold, variance_threshold)

    normalized_data = normalized_data.loc[normalized_data["Trial"] == trial,:].copy()
    original_data = original_data.loc[original_data["Trial"] == trial,:].copy()

    test_data = prep._normalize_data(input_data, False, True, False)
    test_data = test_data.loc[test_data["Trial"] == trial,:].copy()

    
    # 그래프 설정
    fig, axes = plt.subplots(3, 4, figsize=(18, 10))
    axes = axes.ravel()

    relevant_channel_info = prep._relevant_channels_info.get(trial, [])
    relevant_channel = list(map(lambda x: x["channel"], relevant_channel_info))
    
    for i, channel in enumerate(prep._channels):  # Ch1 ~ Ch12
        axes[i].plot(
            test_data['elapsed_time'],
            test_data[channel],
            color='lightpink',
            label='Teest Data'
        )
        axes[i].plot(
            original_data['elapsed_time'],
            original_data[channel],
            color='grey',
            label='Original Data'
        )
        axes[i].plot(
            normalized_data['elapsed_time'],
            normalized_data[channel],
            color='blue',
            label='Normalized Data'
        )

        if channel in relevant_channel:
            pressed_intervals = next((x["pressed_intervals"] for x in relevant_channel_info if x["channel"] == channel), [])
            print(channel, pressed_intervals)
            for pressed_interval in pressed_intervals:
                start, end = pressed_interval
                axes[i].axvspan(
                    original_data['elapsed_time'].iloc[start],
                    original_data['elapsed_time'].iloc[end],
                    color='lightblue', alpha=0.3
                )

        axes[i].set_title(f'Channel {i + 1}')  # Channel 인덱스 수정
        axes[i].set_xlabel('Elapsed Time')
        axes[i].set_ylabel('Normalized Value')
        axes[i].grid()
    
    plt.tight_layout()
    plt.subplots_adjust(hspace=0.4, wspace=0.2)  # 서브플롯 간 간격 조정
    plt.show()
```
